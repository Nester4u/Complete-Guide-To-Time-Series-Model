```python
df[['lat','long','odometer',]]=df[['lat','long','odometer']].fillna(df[['lat','long','odometer']].median())
```

```python
df.isnull().sum()
```

```python
df.county=df.county.fillna('tx')
```

```python
df['image_url']=df['image_url'].ffill()
```

```python
df['description']=df['description'].bfill()
```

```python
df['size']=df['size'].ffill()
```

```python
df['type']=df['type'].ffill()
```

```python
df['paint_color']=df['paint_color'].bfill()
```

```python
df['drive'] = df['drive'].fillna(df['drive'].mode()[0])
```

```python
df['VIN'] = df['VIN'].fillna(df['VIN'].mode()[0])
```

```python
df['transmission'] = df['transmission'].fillna(df['transmission'].mode()[0])
```
